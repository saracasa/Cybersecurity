```python

colab_path = "/content/"
tar_path = colab_path + "diabetes-data.tar.Z"
data_folder = colab_path + "Diabetes-Data"
csv_path = os.path.join(colab_path, "diabetes_combined.csv")


if ("diabetes-data.tar.Z" not in os.listdir(colab_path)):
  print ("Not such file")
else:
  extract_path = colab_path
  !mkdir -p {extract_path} > /dev/null 2>&1
  !tar -xvzf {tar_path} -C {extract_path} > /dev/null 2>&1 || tar -xvf {tar_path} -C {extract_path} > /dev/null 2>&1
  print ("Data correctly extracted")


  if ("diabetes_combined.csv" not in os.listdir(colab_path)):
      # Leggi tutti i file che iniziano con "data-"
      files = sorted([os.path.join(data_folder, f) for f in os.listdir(data_folder) if f.startswith("data-")])
      # Combina in un solo DataFrame
      real_data_list = [pd.read_csv(f, sep="\t", header=None) for f in files]
      real_data = pd.concat(real_data_list, ignore_index=True)
      # Rinominare le colonne
      real_data.columns = ["date", "time", "feature1", "feature2"]
      # Converti le feature numeriche
      real_data["feature1"] = pd.to_numeric(real_data["feature1"], errors='coerce')
      real_data["feature2"] = pd.to_numeric(real_data["feature2"], errors='coerce')
      # Salva come CSV unico
      real_data.to_csv(csv_path, index=False)
      print("CSV salvato in:", csv_path)
      print("Dimensione totale:", real_data.shape)
  else:
      # Leggi CSV gi√† combinato
      real_data = pd.read_csv(csv_path)
      print("Dimensione totale:", real_data.shape)
      print("Data already extracted")
```
